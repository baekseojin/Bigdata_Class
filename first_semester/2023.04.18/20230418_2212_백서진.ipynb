```python
import pandas as pd
import numpy as np
exam = pd.read_csv("./exam_sample.csv", encoding='utf-8')
exam.head
```

```python
df1 = exam.groupby(['class'])
df1
df1.get_group('A')
```

```python
exam.groupby('class').mean()
df1.mean()
```

```python
exam.groupby( [ 'class', 'sex' ] ).mean()
```

```python
exam['math'].groupby(exam['class']).mean()
```

```python
exam.groupby('class')['math'].mean()
```

```python
# agg(새롭게 만들어낼 컬럼명 = )

exam.groupby('class').agg(  mean_math = ( 'math', 'mean' ) )
```

```python
# 수학 평균 / 영어 중앙값 / 반별 학생수

exam.groupby('class').agg( mean_math = ('math','mean'),
                          median_english = ('english', 'median'),
                          n = ('class', 'count') )
```

```python
exam.groupby('class').agg( [ 'mean', 'median', 'count'] )
```

```python
exam.groupby('class').agg({ 'math':['min','max'] , 'english':'mean'  } )
```

```python
df1.apply(lambda x : x.mean() >= 50 )
df1.filter(lambda x : x.math.mean() >= 50 )
```

```python
import pandas as pd
import numpy as np
mpg = pd.read_csv("./mpg.csv", encoding='utf-8')
```

```python
mpg.groupby('category')['cty'].mean().sort_values(ascending=False)
mpg.groupby('category').agg( mean_cty = ('cty', 'mean') ).sort_values('mean_cty', ascending=False)


mpg.groupby('manufacturer')['hwy'].mean().sort_values(ascending=False).head(3)
mpg.groupby('manufacturer').agg( mean_hwy = ('hwy', 'mean') ).sort_values('mean_hwy', ascending=False).head(3)

mpg[ mpg.category == 'compact'].groupby('manufacturer').agg(n=('manufacturer','count')).sort_values('n', ascending=False)
mpg[ mpg.category == 'compact'].value_counts('manufacturer')
```
