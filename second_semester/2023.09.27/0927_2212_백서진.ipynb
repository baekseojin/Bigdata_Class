```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
gender=["F","F","F","M","M","M"]
wei = [65,66,69,67,68,72]
hei = [171,172,176,173,177,178]
age=[23,24,38,43,40,42]

#df    

df = pd.DataFrame({"gender":gender,"wei":wei,"hei":hei, 'age':age })
print(df)

cor = df.corr(method='pearson')
print( cor )

sns.set( style = 'white')
f, ax = plt.subplots(figsize=(12, 12))
sns.heatmap( cor, center=0.5, square=True, linewidths=0.5, cbar_kws={'shrink':0.75}, annot=True )
plt.title('test', size=30)
ax.set_xticklabels( list(df.columns[1:]), size=15, rotation=90 )
ax.set_yticklabels( list(df.columns[1:]), size=15, rotation=90 )
mask = np.zeros_like( cor, dtype=np.bool)
mask[ np.triu_indices_from(mask) ] = True
print( mask )
sns.set( style = 'white')
f, ax = plt.subplots(figsize=(12, 12))

sns.heatmap( cor, mask=mask, center=0.5, square=True, linewidths=0.5, cbar_kws={'shrink':0.75}, annot=True )
plt.title('test', size=30)
ax.set_xticklabels( list(df.columns[1:]), size=15, rotation=90 )
ax.set_yticklabels( list(df.columns[1:]), size=15, rotation=90 )

sns.set( style = 'white')
f, ax = plt.subplots(figsize=(12, 12))
sns.heatmap( cor, center=0.5, square=True, linewidths=0.5, cbar_kws={'shrink':0.75}, annot=True )


plt.title('test', size=30)
ax.set_xticklabels( list(df.columns[1:]), size=15, rotation=90 )
ax.set_yticklabels( list(df.columns[1:]), size=15, rotation=90 )
mask = np.zeros_like( cor, dtype=np.bool)
mask[ np.triu_indices_from(mask) ] = True
print( mask )
sns.set( style = 'white')
f, ax = plt.subplots(figsize=(12, 12))
sns.heatmap( cor, mask=mask, center=0.5, square=True, linewidths=0.5, cbar_kws={'shrink':0.75}, annot=True )
plt.title('test', size=30)
ax.set_xticklabels( list(df.columns[1:]), size=15, rotation=90 )
ax.set_yticklabels( list(df.columns[1:]), size=15, rotation=90 )

```
