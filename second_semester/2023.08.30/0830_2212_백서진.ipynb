```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("./health.csv")
data.head()
```

```python
# 성별 빈도표

gender_freq = pd.DataFrame( data.성별.value_counts())
gender_freq.index = [ 'male', 'female' ]
gender_freq.columns = ['freq']
gender_freq['ratio'] = np.round( gender_freq.freq / sum(gender_freq.freq), 2)
print(gender_freq)
```

```python
# 키 빈도표
height_freq = pd.DataFrame( data.키.value_counts() )
height_freq.columns = ['freq']
height_freq['ratio'] = np.round( height_freq.freq / sum(height_freq.freq), 2 )
print( height_freq )

height_freq = height_freq.sort_index()
```

```python
# 나이 빈도

bins = [ 130, 140, 150, 160, 170, 180, 190, 200 ]
label = ['130대', '140대', '150대', '160대', '170대', '180대', '190대']
height_cut = pd.cut( data.키, bins, right = True, labels=label)
height_class = pd.value_counts( height_cut)
height_class = pd.DataFrame( height_class )
height_class.columns = ['freq']
height_class['ratio'] = np.round( height_class.freq / sum(height_class.freq), 2 )
height_class =  height_class.sort_index()
print( height_class )
```

```python
height_class = pd.merge(height_class, height_class.cumsum(),how='left')
height_class
```

```python
height_class['cum_freq'] = height_class.cumsum().freq
height_class['cum_ratio'] = height_class.cumsum().ratio
```

```python
a = np.array([12, 16, 10, 15, 13, 18])
b = np.array([10, 11, 14, 17, 15, 17])
a_std = np.std(a) # a의 표준편차
b_std = np.std(b) # b의 표준편차
print('A학생의 표준편차 : %.2f, B학생의 표준편차 : %.2f' %(a_std, b_std))
print('더 고른 성적의 학생 :', end=' ')
if a_std < b_std : print('A학생')
elif a_std > b_std : print('B학생')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("kimbab.csv", delimiter='\t')

sajang_data = data['사장님']
cheolsu_data = data['철수']

sajang_hist, sajang_edges = np.histogram(sajang_data, bins=10)
cheolsu_hist, cheolsu_edges = np.histogram(cheolsu_data, bins=10)

plt.bar(sajang_edges[:-1], sajang_hist, width=np.diff(sajang_edges), align='edge', label='Sajang')
plt.bar(cheolsu_edges[:-1], cheolsu_hist, width=np.diff(cheolsu_edges), align='edge', label='Cheolsu', alpha=0.7)

plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.legend()  # 범례 추가
plt.show()


print("-----------------------------------")


mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)

print("평균:", mean)
print("분산:", variance)
print("표준편차:", std_dev)
```
